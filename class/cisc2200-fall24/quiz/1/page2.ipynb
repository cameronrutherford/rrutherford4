## Per-question results

### Question 0 - preparing data

We will be reading the data here, and processing later..

```python
import csv
import numpy as np

results = []

with open('page2.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=",")
    for row in csv_reader:
        results.append(list(map(int, row)))

res_array = np.array(results)

print(f"Question 1: c) Scores {res_array[:,:].tolist()}")  
```

```python
import matplotlib.pyplot as plt

def print_stats(A, bins = [-1,0.1,1.1,2.1,3.1,4.1,5.1]):
    print(f"Mean: {np.mean(A)}")
    print(f"Median: {np.median(A)}")
    vals, counts = np.unique(A, return_counts=True)
    print(f"Mode: {A[np.argmax(counts)]} With count: {counts[np.argmax(counts)]}")
    plt.hist(A, bins = bins, align='right')
```

### Question 2

#### c. 

```python
print_stats(res_array[:,:1])
```

### Question 3

```python
print_stats(res_array[:,1:2])
```

## Page results

```python
#| label: fig-page-2-breakdown
#| fig-cap: >
#|    Results for each question on page 2

print_stats(res_array)
```

```python
#| label: fig-page-2-totals
#| fig-cap: >
#|    Totals for page 2

print_stats(np.sum(res_array, axis=1), bins = [x + 0.1 for x in range(9,16)])
```

### After scaling

```python
#| label: fig-page-2-scaled
#| fig-cap: >
#|    Totals for page 2 after scaling

f = lambda x: x + (15 - x) / 2

print_stats(f(np.sum(res_array, axis=1)), bins = [x + 0.1 for x in range(9,16)])
```
